```python
from crewai import LLM
llm=LLM(model="openai/gpt-4.1-mini-2025-04-14")
```

```python
from crewai.tools import BaseTool
from pydantic import Field


class SQLTool(BaseTool):
    name: str = "sql_tool"
    description: str = "A tool for running SQL queries against the database. Use this to execute SELECT statements to retrieve data, INSERT to add records, UPDATE to modify existing data, DELETE to remove records, and other SQL operations. Provide the complete SQL query as input."
    
    db: any = Field(default_factory=lambda: db)
    
    def _run(self, query: str) -> str:
        """Execute the SQL query and return results"""
        try:
            return str(self.db.run(query))
        except Exception as e:
            return f"Error performing SQL query: {str(e)}"
```

```python
SQLTool().run("select * from __UserSettings")
```

```python
from crewai import Agent, Task, Crew, Process
```

```python
import pandas as pd
import seaborn as sns
```

```python
agent_1 = Agent(
   role="ERP SQL Database Assistant",
   goal="Understand user queries in natural language and provide accurate data insights that match ERP report results by analyzing MariaDB schema and executing appropriate SQL queries with proper business logic and filtering",
   backstory="You are an experienced ERP data analyst who specializes in translating business questions into SQL queries for MariaDB-based ERP systems. You have deep knowledge of ERP database structures, understand how ERP reports apply business rules and filters, and can quickly identify table relationships, column meanings, and data patterns. You ensure your SQL results match exactly what users see in their ERP reports by applying the same filtering logic, date contexts, and business rules. You communicate findings clearly in natural language while being precise about the data you're presenting and always cross-reference against expected ERP report behavior.",
   tools=[SQLTool()],
   llm=llm,
   verbose=True
)

task_1 = Task(
   description="Analyze the user's natural language query {query} and generate SQL that produces results identical to what they would see in their ERP reports. First explore the database schema to understand available tables and columns, then consider what business filters and logic the ERP system applies by default (such as enabled/disabled status, active records, date ranges). Generate and execute appropriate SQL queries that include proper joins, filtering, and calculations to match ERP report standards. Present the results in a clear, conversational manner that aligns with how data appears in the ERP interface. Handle follow-up questions and maintain context throughout the conversation while ensuring consistency with ERP report behavior.",
   expected_output="A natural language response that directly answers the user's question with relevant data insights that match ERP report results, including any clarifications about the data source, business logic applied, limitations, or additional context that would be helpful for ERP users.",
   agent=agent_1
)
```

```python
agent_2 = Agent(
   role="Data Conversion Specialist",
   goal="Convert SQL query results and data insights from the ERP SQL Database Assistant into structured pandas DataFrames for further analysis and processing, and export as CSV",
   backstory="You are a data processing specialist who takes the output from ERP database queries and converts them into pandas DataFrames. You understand how to parse SQL results, maintain data integrity during conversion, and ensure that the DataFrame structure preserves all relevant information from the original query results. You work closely with database analysts to transform their findings into a format suitable for data analysis workflows and export the final results as CSV files for easy sharing and further use.",
   llm=llm,
   allow_code_execution=True,
   verbose=True
)

task_2 = Task(
   description="Take the output from agent_1 (ERP SQL Database Assistant) and convert it into a pandas DataFrame. Parse the SQL query results, maintain proper data types, column names, and row structure. Ensure the DataFrame accurately represents the data that was retrieved from the ERP database query. Handle any data formatting or type conversion needed to create a clean, usable DataFrame structure. After creating the DataFrame, export it as a CSV file for easy access and sharing.",
   expected_output="A pandas DataFrame containing the converted data from agent_1's SQL query results, with proper column headers, data types, and complete data integrity preserved from the original query output, along with a CSV file export of the DataFrame.",
   agent=agent_2,
   context=[task_1]
)
```

```python
agent_3 = Agent(
   role="Data Visualization Specialist",
   goal="Create insightful and professional seaborn visualizations using pandas DataFrames to help users understand ERP data patterns, trends, and insights",
   backstory="You are a data visualization expert who specializes in creating meaningful charts and plots using seaborn and matplotlib. You understand how to choose the right visualization types based on data characteristics, create clear and informative plots that tell a story, and ensure visualizations are properly formatted with appropriate titles, labels, and styling. You work with clean pandas DataFrames to generate publication-ready visualizations that help business users quickly grasp data insights from their ERP systems.",
   llm=llm,
   allow_code_execution=True,
   verbose=True
)

task_3 = Task(
   description="Take the pandas DataFrame output from agent_2 and create appropriate seaborn visualizations based on the data characteristics and business context. Analyze the DataFrame structure to determine the most suitable plot types (bar plots, line charts, heatmaps, scatter plots, etc.). Generate Python code for clean, professional visualizations with proper titles, axis labels, legends, and formatting. Consider the business meaning of the data when choosing colors, scales, and plot styles. Execute the generated visualization code to create and display the plots, then save them as high-quality image files.",
   expected_output="Professional seaborn visualizations that effectively communicate the ERP data insights, with properly formatted charts including titles, labels, legends, executed and displayed plots, and saved as image files for easy sharing and presentation.",
   agent=agent_3
)
```

```python
crew = Crew(
    agents=[agent_1,agent_2,agent_3],
    tasks=[task_1,task_2,task_3],
    process=Process.sequential
)
```

```python
crew.kickoff({"query":"Which employees have the highest number of leave applications this year?"})
```
